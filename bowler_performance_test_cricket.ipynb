##Ques 1:Remove the columns BBI and BBM.

#Ques 2: How many players played for ICC?

```python
#String spliting 
df_player = df['Player'].str.split("(", expand=True)
display(df_player.head())

#Concatenation of two string
df = pd.concat([df, df_player], axis=1)
df = df.drop('Player', axis=1)

# Renaming the new column
df = df.rename(columns={0: 'Player',
                        1: 'Country'})

#Changing the column order
new_col_sequence = ['Player', 'Span', 'Mat', 'Inns', 'Balls', 'Runs', 'Wkts',
                    'Ave', 'Econ', 'SR', '5', '10','Country']
df = df[new_col_sequence]

#Checking for bowler played for ICC
def icc_check(x):
    if "ICC" in x:
        return "Yes"
    else:
        return "No"
df['played_for_ICC'] = df['Country'].apply(icc_check)
display(df.head())

#Value count 
df['played_for_ICC'].value_counts()
```

#Ans: In this dataset 5 player played for ICC.


#Ques 3:How many different countries are present in this dataset? 

```python
#1st removing the "ICC/" and '/ICC' from country to count number of country present in dataset
df['Country'] = df['Country'].str.replace("ICC/", "")
df['Country'] = df['Country'].str.replace("/ICC", "")

#Removing ')' from country #new_keys = df['keys'].str.rstrip('/')
df['Country'] = df['Country'].str.rstrip(')')
display(df.head())

#Now need to check the uniqe value in country column
df['Country'].value_counts()
```

#Answer: Here, in country column 10 unique value are present. Number of country playing test cricket is 10 in this dataset.

#Ques 4: Which player(s) had played for the longest period of time?

```python
#Spilting the string value of 'Span' into two column
df['start_year'] = df['Span'].str[0:4]
df['end_year'] = df['Span'].str[5:]

#Drop the "Span" column
df = df.drop("Span", axis=1)
display(df.head())

#Changing the data type from 'str' to 'int'
df['start_year'] = df['start_year'].astype('int') 
df['end_year'] = df['end_year'].astype('int')
print(df.info())

df['years_played'] = df['end_year'] - df['start_year']
df = df.drop(['start_year', "end_year"], axis=1)
display(df.head())
```

```python
#storing the column 'year_played' as ascending order
df.sort_values(by='years_played', ascending = False).head()
```

#Answer 4: Imran Khan had played for 21 years which is the longest period as a bolower in the dataset.

#Ques 5: Which player(s) had played for the shortest period of time?

```python
#Sort the year_played column in descending order
df.sort_values(by='years_played', ascending = True).head()
```

#Answer 5: GP Swann had played for 5 years which is the shortest period as a bolower in this dataset.
